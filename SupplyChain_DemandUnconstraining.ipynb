```python
# strategy:
# create a big loop that takes a censored curve as input and fills in the nan's
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.api import ExponentialSmoothing
from itertools import product

# write a function that takes alpha,beta,series as input
# and produces the traning sample MSE
# allow 3 periods for burn-in
def DESMSE(alpha,beta,series):
    return mean_squared_error(series[3:],
                              ExponentialSmoothing(series,trend="add", initialization_method='estimated').fit(smoothing_level = alpha,smoothing_trend = beta).predict(3,series.size-1))

# the function below allows you to construct a dataframe of alpha-beta combinations
def expand_grid(dictionary):
   return pd.DataFrame([row for row in product(*dictionary.values())],
                       columns=dictionary.keys())

# define the alpha and beta values to try as a dictionary
dictionary = {'alpha': np.arange(0.1, 1.1, 0.1),
              'beta': np.arange(0.1, 1.1, 0.1)}


# predict curve by curve
# note: the shape attribute for numpy arrays returns the dimensions of the array. If a df has n rows and m columns, then df.shape is (n,m). So Y.shape[0] is n
for row_number in np.arange(0,df_des.shape[0]):
    # take out a censored curve, drop nan values
    temp = df_des.iloc[row_number,:].dropna().reset_index(drop=True)

    # convert cumulative booking into daily booking
    # .diff() will work too
    # temp[1:] gets booking vlaues for days 2-16, temp[:-1] gets booking vlaues for days 1-15,
    temp = np.append(temp[0],temp[1:].values - temp[:-1].values)

    # construct alpha,beta values to try
    des_results = expand_grid(dictionary)

    # add a column called mse, as a place holder for putting in the MSE results later
    des_results["mse"] = np.repeat(0,len(des_results))

    # reuse the DESMSE function above to cycle through all alpha beta values
    # (alpha,beta,series) = (alpha,beta,temp)
    for alpha in np.arange(0.1, 1.1, 0.1):
        for beta in np.arange(0.1, 1.1, 0.1):
           des_results.loc[(des_results["alpha"] == alpha) & (des_results["beta"] == beta),"mse"] = DESMSE(alpha,beta,temp)

    # find the best (alpha beta): sort and put on row 1
    des_results = des_results.sort_values("mse")

    # use the best (alpha,beta) to forecast for the censored days
    # post results onto df_des
    # forecast(16-temp.size) ensures that we forecast the censored days
    predicted_daily_arrival = ExponentialSmoothing(temp,trend="add", initialization_method='estimated'). \
                                         fit(smoothing_level = des_results.iloc[0,0],
                                             smoothing_trend = des_results.iloc[0,1]). \
                                         forecast(16-temp.size)

    # cumsum the daily arrival and add back to the last value that's not nan
    df_des.iloc[row_number,temp.size:] = predicted_daily_arrival.cumsum() + df_des.iloc[row_number,temp.size-1]
```

### b. Averaging method estimations

```python
grocery_2 = grocery_old.copy()
```

```python
grocery_2 = grocery_2.drop(['censored','des','Total'],axis=1)
```

```python
grocery_2.columns = np.arange(1,17)
grocery_2 = pd.melt(grocery_2.reset_index(),id_vars='index',var_name = 'hour',value_name='grocery')
# sort by curve and day
grocery_2 = grocery_2.sort_values(by=['index','hour'])
# ensure conversion done as expected
grocery_2
```

```python
grocery_2['censor'] = grocery_2['grocery'].shift(-1).isna() | grocery_2['grocery'].isna()
grocery_2
```

这行代码执行了几个操作：

1. `grocery_2['grocery'].shift(-1)`：这部分代码使用了 `shift` 函数来将 `grocery` 列向上移动一行。参数 `-1` 表示向上移动一行，因此这一步将 `grocery` 列中的值向上移动了一行，导致每个值都变成了其下一行的值。

2. `.isna()`：这是 Pandas Series 对象的一个方法，它返回一个布尔值的 Series，表示该 Series 中的每个值是否为缺失值（NaN）。如果值是 NaN，则返回 True，否则返回 False。

3. `grocery_2['grocery'].shift(-1).isna()`：这一部分将步骤1和步骤2组合起来，首先将 `grocery` 列向上移动一行，然后检查每个值是否为 NaN，最终得到一个布尔值的 Series。

这行代码的目的是创建一个布尔值的 Series，该 Series 中的每个值表示相应行的 `grocery` 列是否为缺失值，并且该行的下一行是否为 NaN。

### c. Comparing DES and Averaging method results

Comparing the results of the AVERAGING and DES methods from the above table, it's evident that the averaging method is "less aggressive" than the DES method. This indicates that the averaging method, with its simplicity and conservative nature, tends to produce smoother but potentially less accurate forecasts.

The averaging method generates forecasts by computing the average of historical values over a specified period. This approach assumes that future values will resemble the historical average and does not involve intricate smoothing techniques or trend adjustments. Consequently, it yields forecasts that change gradually over time.

In contrast, the DES method incorporates both level and trend components through exponential smoothing. It assigns varying weights to historical observations, giving more importance to recent data points. Moreover, it models the trend in the data, enabling it to capture both short-term fluctuations and long-term trends. Consequently, DES produces forecasts that can respond more rapidly to changes in the data.

Regarding unconstrained prediction, the averaging method tends to be less aggressive due to its conservative reliance on historical averages. It adapts less dynamically to changes compared to DES, leading to smoother and more stable forecasts over time.
